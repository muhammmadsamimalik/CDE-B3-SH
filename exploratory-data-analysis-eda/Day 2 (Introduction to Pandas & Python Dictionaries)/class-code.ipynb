```python
df = pd.DataFrame({"A": [1, 2], "B": [3, 4]})

# Renaming columns and creating a new DataFrame (default behavior)
new_df = df.rename(columns={"A": "Column_A_New"})
print("Original DataFrame:")
print(df)
print("\nNew DataFrame after rename:")
print(new_df)
```

```python
new_df.loc[0, 'Column_A_New'] = 99
print("\nNew DataFrame after modification:")
print(new_df)
print("\nOriginal DataFrame (should be unchanged):")
print(df)
```

```python
df = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6]})
# The rename with copy=False creates a new df_renamed object,
# but shares the same underlying data as df.
df_renamed = df.rename(columns={'A': 'C'}, copy=False)

# Modifying the data through the renamed DataFrame would also modify the original.
df_renamed.iloc[0, 0] = 999

# In older pandas, this would print 999 in both DataFrames.
print("Original df (pre-CoW behavior):\n", df)
print("\nRenamed df (pre-CoW behavior):\n", df_renamed)
```

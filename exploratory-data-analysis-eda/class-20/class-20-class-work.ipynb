### Merging Tables

![join](./img/joins.png)

## Inner Join

![join](./img/inner_join.png)

```python
left_dict = {
    'Date': ['1/1/2020', '1/2/2020', '1/3/2020', '1/4/2020'],
    'CountryID': [1, 1, 3, 2],
    'Units': [40, 25, 30, 35]
}
```

```python
left_table = pd.DataFrame(left_dict)
left_table
```

```python
right_dict = {
    'CountryID': [3, 4],
    'Country': ['Panama', 'Spain']
}
```

```python
right_table = pd.DataFrame(right_dict)
right_table
```

```python
left_table.merge(right_table, left_on='CountryID', right_on='CountryID')
```

```python
right_dict = {
    'ID': [3, 4],
    'Country': ['Panama', 'Spain']
}
```

```python
right_table = pd.DataFrame(right_dict)
right_table
```

```python
left_table.merge(right_table, left_on='CountryID', right_on='ID')
```

```python
left_table.merge(right_table, left_on='CountryID', right_on='ID').drop('ID', axis=1)
```

## FULL/OUTER JOIN 

![full join](./img/full_outer_join.png)

```python
left_dict = {
    'Date': ['1/1/2020', '1/2/2020', '1/3/2020', '1/4/2020'],
    'CountryID': [1, 1, 3, 2],
    'Units': [40, 25, 30, 35]
}

left_table = pd.DataFrame(left_dict)
left_table
```

```python
right_dict = {
    'ID': [1, 2, 3, 4],
    'Country': ['USA', 'Canada', 'Panama', 'Pakistan']
}

right_table = pd.DataFrame(right_dict)
right_table
```

## Left Anti Join

![left anti](./img/left_anti_join.png)

```python
left_dict = {
    'Date': ['1/1/2020', '1/2/2020', '1/3/2020', '1/4/2020'],
    'CountryID': [1, 1, 3, 2],
    'Units': [40, 25, 30, 35]
}

left_table = pd.DataFrame(left_dict)
left_table
```

```python
right_dict = {
    'CountryID': [3, 4],
    'Country': ['USA', 'Canada']
}

right_table = pd.DataFrame(right_dict)
right_table
```

```python
left_join = left_table.merge(right_table, left_on='CountryID', right_on='CountryID', how='left', indicator=True)
left_join
```

```python
left_join[left_join['_merge'] == 'left_only']
```

## Right Anti Join

![right anti join](./img/right_anti_join.png)

```python
left_dict = {
    'Date': ['1/1/2020', '1/2/2020', '1/3/2020', '1/4/2020'],
    'CountryID': [1, 1, 3, 2],
    'Units': [40, 25, 30, 35]
}

left_table = pd.DataFrame(left_dict)
left_table
```

```python
right_dict = {
    'CountryID': [3, 4],
    'Country': ['USA', 'Panama']
}

right_table = pd.DataFrame(right_dict)
right_table
```

```python
right_join = left_table.merge(right_table, left_on='CountryID', right_on='CountryID', indicator=True, how='right')
right_join
```

```python
right_join[right_join['_merge'] == 'right_only']
```

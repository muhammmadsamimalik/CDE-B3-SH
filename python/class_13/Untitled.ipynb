### Task:
Create an abstract class Notification:
- It should have an abstract method **send(message)** that accepts a message as a  parameter.
- It should also have a regular method **show_info()** that prints "Sending Notification".

Create two classes that inherit from Notification
- 1. EmailNoti -> should implement send() to print "Sending Email: <message>"
  2. SMSNoti -> should implement send() to print "Sending SMS: <message>"

```python
class Notification:
    def send(self, message):
        pass

    def show_info(self):
        print('Sending Notification...')

class EmailNotification(Notification):
    def send(self, message):
        print(f"Sending Email: {message}")

class SMSNotification(Notification):
    def send(self, message):
        print(f"Sending SMS: {message}")
```

```python
email_obj = EmailNotification()
sms_obj = SMSNotification()

email_obj.show_info()
email_obj.send("Welcome to DE class")
sms_obj.show_info()
sms_obj.send("Welcome to CDE class")
```

### Task:
- create a class called StudentLogin
  - initailize two private attributes (username, password)
  - create 1 private method called dbconnection which validate your private attributes
  - create a login method which take two parameters (username, password) and execute the dbconnection method if both are valid print a valid message otherwise print invalid user.

```python
class StudentLogin:
    def __init__(self):
        self.__username = "umair"
        self.__password = 1234

    def __dbconnection(self, username, password):
        if self.__username == username and self.__password == password:
            print("Valid User")
        else:
            print("Invalid User")

    def login(self, username, password):
        self.__dbconnection(username, password)
```

```python
std1 = StudentLogin()
```

```python
std1.login("ali", 1234)
```

```python
std1.login("umair", 1234)
```

### Data File Handling
- write -> w
- Read -> r
- append -> a
- write/read w+

```python
# Method 1
f = open("test.txt", "w")
f.write("This is the file content.")
f.close()
```

```python
# Method 2
with open("test1.txt", "w") as f:
    f.write("This is my new file.")
```

```python
with open("test1.txt", "a") as f:
    f.write("This is the appended content.")
```

```python
with open("test1.txt", "a") as f:
    f.write("\nThis is a new line.")
```

```python
with open("test1.txt", 'r') as f:
    print(f.read())
```

```python
with open("test2.txt", 'a+') as f:
    f.write("\nThis is the new conten.")
    f.seek(0)
    content = f.read()
    print(content)
```

### List and Dict Comprehension

```python
nums = []

for i in range(1, 11):
    if i % 2 == 0:
        nums.append(i)

nums
```

```python
nums = [i for i in range(1, 11) if i % 2 == 0]
nums
```

```python
nums = []

for i in range(1, 11):
    if i % 2 == 0:
        nums.append("even")
    else:
        nums.append("odd")

nums
```

```python
nums = ["even" if i % 2 == 0 else "odd" for i in range(1, 11)]
nums
```

```python
nums_dict = {i: "even" if i % 2 == 0 else "odd" for i in range(1, 11)}
nums_dict
```

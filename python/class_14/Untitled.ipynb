### File Handling

```python
file = open('demo.txt', 'w')
file.write('This is a demo file.')
file.close()
```

```python
with open('test.txt', 'w') as f: # mode 'w' to write something
    f.write("This is a text file.")
```

```python
with open('test.txt', 'a') as f: # mode 'a' to append content in the file
    f.write('\nThis is a new line.')
```

```python
with open('test.txt', 'r') as f: # mode 'r' to read the content from file
    content = f.read()
    print(content)
```

### CSV File Handling

```python
import csv
```

```python
with open('csv_test.csv', 'w') as f:
    csv_data = csv.writer(f)
    csv_data.writerow(["Name", "F. Name"])
```

```python
with open('csv_test.csv', 'r') as f:
    csv_data = csv.reader(f)
    lst = []
    for i in csv_data:
        lst.append(i)
```

```python
nums = []
for i in lst[1:]:
    int_nums = int(i[1])
    nums.append(int_nums)
```

```python
sum(nums)/len(nums)
```

```python
with open('csv_test.csv', 'a', newline="") as file:
    # file.write('\n')
    csv_data = csv.writer(file)
    csv_data.writerow(["Salman", '48'])
```

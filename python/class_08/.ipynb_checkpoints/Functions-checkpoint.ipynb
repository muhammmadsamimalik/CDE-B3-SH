```python
def bio_data(first_name, last_name, course, **other_data):
    print("First Name: ", first_name)
    print("Last Name: ", last_name)
    print("Course: ", course)
    for i, j in other_data.items():
        print(i, j)
```

### Task:
- write a function calle Registration which take email, name and password from user and save them in dict of dict, for the first dict the key should be user's email and value should be a dict where you will store user's name and password.
- write another function which take email and password and check in the saved data, if emial matches the key so check his password and print  a login message with his name.

```python
user_info = {}
```

```python
def registration(name, email, password):
    user_info[email] = {'name': name, 'password': password}
```

```python
registration('Ali', 'ali@email.com', 1234)
```

```python
user_info
```

```python
registration('Umair', 'umair@email.com', 12345)
```

```python
user_info
```

```python
def login(email, password):
    if email in user_info.keys():
        if password == user_info[email]['password']:
            print(f'{user_info[email]['name']}, Login Successful.')
```

```python
login('ali@email.com', 1234)
```

### Task
    Create a Dictionary called 'company' with 3 different departments (HR, IT, Finance), create a list for each department and store two employee details like name, salary, and experience using a dictionary in each department.

- Print all employees in IT department
- Add a new employee 'Qasim' with salary 55000 and experiecne 1 to HR
- Increase salary of Qasim by 10000.
- Print all employees who have experience more than 3 years (any department).

```python
company = {
    "HR": [
        {
            "name": "Ali", "salary": 40000, "experience": 3
        },
        {
            "name": "Sara", "salary": 30000, "experience": 2
        }
    ],
    "IT": [
        {
            "name": "Umair", "salary": 20000, "experience": 3
        },
        {
            "name": "Ahmed", "salary": 70000, "experience": 5
        }
    ],
    "Finance": [
        {
            "name": "Zara", "salary": 90000, "experience": 6
        },
        {
            "name": "Mehmood", "salary": 80000, "experience": 4
        }
    ]
}
```

```python
for i in company["IT"]:
    print(i)
```

```python
company["HR"].append({'name': 'Qasim', 'salary': 55000, 'experience': 1})
```

```python
company["HR"]
```

```python
company["HR"][2]["salary"] = company["HR"][2]["salary"] + 10000
company["HR"]
```

```python
company["HR"][2]["salary"]
```

```python
for i in company:
    for j in company[i]:
        if j["experience"] > 3:
            print(j)
```

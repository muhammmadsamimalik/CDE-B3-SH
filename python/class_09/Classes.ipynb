```python
class Cars:
    def __init__(self, model, color, make_year): # --> Constructor
        self.model = model
        self.color = color
        self.mk_year = make_year

    def start():
        print("Car can start.")

    def car_intro(self, doors_count):
        print(f'The car model is {self.model}, {doors_count} Doors, Car Color: {self.color}')
```

```python
civic = Cars("Honda", "Black", 2024)
```

```python
civic.car_intro(4)
```

### Task:
- Create a student class with 3 attributes (name, age, roll_number) and with 2 methods (Intro and Is_Present)

```python
class Student:
    def __init__(self, name, age, roll_number):
        self.name = name
        self.age = age
        self.roll_num = roll_number

    def intro(self):
        print(f'Hi, I am {self.name}, {self.age} years old, Roll Number: {self.roll_num}')

    def is_present(self, attendence):
        print(f'{self.name} is {attendence}')
```

```python
student1 = Student("Umair", 20, 123)
```

```python
student1.intro()
```

```python
student1.is_present("Absent")
```

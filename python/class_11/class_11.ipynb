```python
a = [[1, 2, 3], [4, 5, 6]]

# Creating a shallow copy of the nested list 'original'
b = copy.copy(a)
```

### Recursion

```python
def rec(num):
    if num == 1:
        return 1
    add = num + rec(num - 1)
    return add
```

```python
rec(15)
```

### Lambda Function
- Function with any name (anonymous function)
- Never been called before and after
- No use of return keyword

lambda arguments: expression

```python
def addition(x, y):
    return x + y
```

```python
addition(4, 6)
```

```python
(lambda x, y: x + y)(4, 5)
```

```python
def f(x):
    if x % 2 == 0:
        return x

f(4)
```

```python
f = lambda x: x**2

f(4)
```

```python
f = lambda x: x*2

f(4)
```

```python
addition = lambda x, y: x + y
```

```python
addition(4, 6)
```

```python
def even_nums():
    for num in range(1, 16):
        if num % 2 == 0:
            print(num)
```

```python
even_nums()
```

```python
nums = [1, 2, 3, 4, 5, 6, 7, 8]

even_func = lambda x: "even" if x % 2 == 0 else "odd"
```

```python
even_func(4)
```

### Map Function
- When you need to run a function on each element of an iteriable or list

```python
lst = list(map(even_func, nums))
lst
```

### Filter function


```python
list(filter(lambda item: item if item == "even" else 0, lst))
```

### Reduce

```python
from functools import reduce
lst = [1,2,3,4,5]
reduce(lambda x , y : x - y , lst)
```

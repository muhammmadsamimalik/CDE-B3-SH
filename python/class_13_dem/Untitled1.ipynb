```python
payment1.pay(500)
```

```python
payment2 = JazzCash()
payment2.pay(790)
```

```python
class StudentLogin:
    def __init__(self):
        self.__username = "Umair"
        self.__password = 12345

    def __dbconnectivity(self, username, password):
        if self.__username == username and self.__password == password:
            print("Valid User")
        else:
            print("Invalid User")

    def login(self, username, password):
        self.__dbconnectivity(username, password)

    def resetpass(self, password):
        self.__password = password
```

```python
std1 = StudentLogin()
```

```python
std1.login("Umair", 1245)
```

```python
std1.resetpass(4321)
```

```python
std1.login("Umair", 4321)
```
